```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
STEM = ['Psychology', 'Biology', 'Math and Statistics',
        'Physical Sciences', 'Computer Science', 'Engineering']
LIBERAL_ARTS = ['Foreign Languages', 'English', 'Communications and Journalism',
               'Art and Performance', 'Social Sciences and History']
OTHER = ['Health Professions', 'Public Administration',
         'Education', 'Agriculture', 'Business', 'Architecture']
```

```python
df = pd.read_csv('./percent-bachelors-degrees-women-usa.csv')
df.head()
```

```python
fig = plt.figure(figsize=(16, 16))

# STEM degrees
for i, col in enumerate(STEM):
    ax = fig.add_subplot(6, 3, i*3 + 1)
    ax.plot(df['Year'], df[col], c='blue')
    ax.plot(df['Year'], 100 - df[col], c='orange')
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.set_title(col)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    # Add text annotation     
    if i == 0:
        ax.text(2005, 85, 'Women')
        ax.text(2007, 10, 'Men')
    elif i == len(STEM) - 1:
        ax.text(2007, 70, 'Men')
        ax.text(2005, 25, 'Women')
    
    # Remove redundancies
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    ax.tick_params(left=False, bottom=False)
    
    if i != len(STEM) - 1:
        ax.set_xticklabels('')
        
# LIBERAL_ARTS degrees
for i, col in enumerate(LIBERAL_ARTS):
    ax = fig.add_subplot(6, 3, i*3 + 2)
    ax.plot(df['Year'], df[col], c='blue')
    ax.plot(df['Year'], 100 - df[col], c='orange')
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.set_title(col)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    # Add text annotation     
    if i == 0:
        ax.text(2005, 75, 'Women')
        ax.text(2007, 20, 'Men')
    elif i == len(LIBERAL_ARTS) - 1:
        ax.text(2007, 60, 'Men')
        ax.text(2005, 35, 'Women')
    
    # Remove redundancies
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    ax.tick_params(left=False, bottom=False)
    
    if i != len(LIBERAL_ARTS) - 1:
        ax.set_xticklabels('')
        
# OTHER degrees
for i, col in enumerate(OTHER):
    ax = fig.add_subplot(6, 3, i*3 + 3)
    ax.plot(df['Year'], df[col], c='blue')
    ax.plot(df['Year'], 100 - df[col], c='orange')
    ax.set_ylim(0, 100)
    ax.set_yticks([0, 100])
    ax.set_title(col)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    
    # Add text annotation     
    if i == 0:
        ax.text(2005, 95, 'Women')
        ax.text(2007, 5, 'Men')
    elif i == len(OTHER) - 1:
        ax.text(2007, 70, 'Men')
        ax.text(2005, 25, 'Women')
    
    # Remove redundancies
    for spine in ax.spines.values():
        spine.set_visible(False)
        
    ax.tick_params(left=False, bottom=False)
    
    if i != len(OTHER) - 1:
        ax.set_xticklabels('')
    
plt.show()
```
