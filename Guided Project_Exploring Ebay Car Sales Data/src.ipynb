### Clean column names

```python
import re
```

```python
def convert(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()
```

```python
col_names = autos.columns.values
col_names
```

```python
new_col_names = []
for col in col_names:
    if col == 'yearOfRegistration':
        col = 'registration_year'
    elif col == 'monthOfRegistration':
        col = 'registration_month'
    elif col == 'notRepairedDamage':
        col = 'unrepaired_damage'
    elif col == 'dateCreated':
        col = 'ad_created'
    else:
        col = convert(col)
    new_col_names.append(col)
    
new_col_names
```

```python
autos.columns = new_col_names
autos.head()
```

- Converted all column names from camelcase to snakecase
- Changed some column names

### Clean data types and representation

```python
autos.describe(include="all")
```

- columns 'abtest', 'gearbox', and 'unrepaired_damage' have only two unique values. Should determine if they must be converted to boolean type
- For 'seller' and 'offer_type', almost all values are the same thus they are not very informative
- 'price' and 'odometer' should be converted to numeric types

```python
autos[['price', 'odometer']].head()
```

```python
price = autos['price']
odometer = autos['odometer']
```

```python
price = price.str.replace('$', "").str.replace(',', '').astype('int')
odometer = odometer.str.replace('km', '').str.replace(',', '').astype('int')
```

```python
autos['price'] = price
autos['odometer'] = odometer
autos.rename(index=str, columns={'odometer': 'odometer_km'}, inplace=True)
autos.info()
```

### Remove Outliers

```python
from scipy import stats
```

```python
pd.Series(np.abs(stats.zscore(autos['price']))).sort_values(ascending=False)
```

```python
autos = autos.loc[np.abs(stats.zscore(autos['price'])) < 3]
```

- Removed values with z-scores higher than or equal to 3 from 'price' column

### Aggregate by Brand

```python
autos['brand'].value_counts()
```

```python
top_20_brands = autos['brand'].value_counts().index[:20]
```

```python
mean_price_by_brand = {}

for brand in top_20_brands:
    mean_price = autos.loc[autos['brand'] == brand]['price'].mean()
    mean_price_by_brand[brand] = mean_price
```

```python
mean_price_by_brand
```

```python
mean_price_series = pd.Series(mean_price_by_brand)
```

```python
agg_by_brand = pd.DataFrame(mean_price_series, columns=['mean_price'])
agg_by_brand
```

```python
mean_mileage_by_brand = {}

for brand in top_20_brands:
    mean_mileage = autos.loc[autos['brand'] == brand]['odometer_km'].mean()
    mean_mileage_by_brand[brand] = mean_mileage
```

```python
mean_mileage_series = pd.Series(mean_mileage_by_brand)
```

```python
agg_by_brand['mean_mileage'] = mean_mileage_series
```
